#### Una vez tenemos todas las columnas de "country" estandarizadas, tenemos que crearnos un dataset que nos aúne todos los nombres de los países con sus respectivos índices. Obviamente, esta columna no podrá tener nombres repetidos de países, ya que será un dataset de referencia, con el que unir los otros 3 dataframes en SQL. 

#### Así, primero debemos crear 3 columnas de country_name (1 por cada tabla), las juntaremos verticalmente y eliminaremos los duplicados (así obtendremos todos los nombres de países que se pudieran encontrar en cualquiera de los tres dataframes). Depués les aplicarmeos un índex (que empiece en 1 y no en 0) y así habremos creado nuestro dataframe de referencia para la DATABASE en SQL.

```python
c1=df_climate.country
```

```python
c2=df_poverty.country
```

```python
c3=df_happiness.country
```

```python
df_country=pd.concat([c1,c2,c3])
```

```python
df_country.shape
```

```python
df_country=df_country.drop_duplicates()
```

Aquí ya tenemos la columna con todos los valores únicos de nombres de país que se encuentren en cualquiera de los 3 dataframes.

```python
df_country.shape
```

```python
df_country=df_country.reset_index()
df_country["country_id"]=df_country.index+1
```

```python
df_country.drop(columns="index",inplace=True)
```

Hasta aquí, ya le hemos añadido el índice (que parte de 1) para asignarle a cada país, un valor. Exportamos el dataframe:

```python
df_country.to_csv("country_id.csv",index=False)
```
